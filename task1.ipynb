# Step 2:

```python
category_counts = df['Categories'].value_counts()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Distribution of Categories among Top YouTube Streamers')
plt.xlabel('Categories')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame
correlation_likes = df['Suscribers'].corr(df['Likes'])
correlation_comments = df['Suscribers'].corr(df['Comments'])

print(f'Correlation between Subscribers and Likes: {correlation_likes:.2f}')
print(f'Correlation between Subscribers and Comments: {correlation_comments:.2f}')
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
plt.figure(figsize=(12, 8))

# Scatter plot for Subscribers vs Likes
plt.subplot(2, 1, 1)
plt.scatter(df['Suscribers'], df['Likes'], alpha=0.5, color='blue')
plt.title('Subscribers vs Likes')
plt.xlabel('Subscribers')
plt.ylabel('Likes')

# Scatter plot for Subscribers vs Comments
plt.subplot(2, 1, 2)
plt.scatter(df['Suscribers'], df['Comments'], alpha=0.5, color='green')
plt.title('Subscribers vs Comments')
plt.xlabel('Subscribers')
plt.ylabel('Comments')

plt.tight_layout()
plt.show()
```

# Step 4:

```python
average_subscribers = df['Suscribers'].mean()
average_visits = df['Visits'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

# Visualize the averages
metrics = ['Suscribers', 'Visits', 'Likes', 'Comments']
average_values = [average_subscribers, average_visits, average_likes, average_comments]

plt.figure(figsize=(12, 8))
sns.barplot(x=metrics, y=average_values, palette='viridis')
plt.title('Average Performance Metrics')
plt.ylabel('Average Count')
plt.show()
```

# Step 5:

```python
# Identify specific categories with exceptional performance metrics
average_metrics_by_category = df.groupby('Categories').agg({
    'Suscribers': 'mean',
    'Visits': 'mean',
    'Likes': 'mean',
    'Comments': 'mean'
}).reset_index()

# Visualize average metrics by category
metrics = ['Suscribers', 'Visits', 'Likes', 'Comments']

plt.figure(figsize=(12, 8))
for metric in metrics:
    sns.barplot(x='Categories', y=metric, data=average_metrics_by_category, palette='viridis')
    plt.title(f'Average {metric} by Content Category')
    plt.xlabel('Categories')
    plt.ylabel(f'Average {metric}')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.show()
```
